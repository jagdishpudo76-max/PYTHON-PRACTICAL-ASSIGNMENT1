```python

def pop_duplicates(my_list):
  for i in range(len(my_list) - 1, -1, -1):
    item = my_list[i]
    if item in my_list[:i]:
      my_list.pop(i)    
  return my_list

#data_list = [10, 20, 10, 30, 40, 20, 50, 40, 10]
data_list = ["apple", "banana", "apple", "orange", "banana"]
print(f"Original List:\n{data_list}")
pop_duplicates(data_list)
print(f"\n\n\nList after sorting:\n{data_list}")
```
