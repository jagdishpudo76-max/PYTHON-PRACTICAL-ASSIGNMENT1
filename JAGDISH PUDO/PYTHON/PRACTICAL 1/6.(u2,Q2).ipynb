```python
num = int(input("Enter a number: "))
```

```python
binary_value = bin(num)[2:]      # remove '0b'
octal_value = oct(num)[2:]       # remove '0o'
hex_value = hex(num)[2:]         # remove '0x'

print("Binary:", binary_value)
print("Octal:", octal_value)
print("Hexadecimal:", hex_value)
```
