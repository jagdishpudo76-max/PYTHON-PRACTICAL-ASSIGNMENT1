```python
import os
files =[f for f in os.listdir('.') if os.path.isfile(f) ]
catergorized_files = {}
for file in files:
    if '.' in file:
        extension = file.split('.')[-1]
    
    else:
    
        extension = ''

    if extension not in catergorized_files:
        catergorized_files[extension] =[]
    catergorized_files[extension].append(file)
print(catergorized_files)
```
