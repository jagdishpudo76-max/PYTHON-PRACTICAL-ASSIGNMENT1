```python
import random
pass1=[]
str="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz%#!*$"
#str1=list(str)
c=0

while(c<12):
    pass1.append(random.choice(str))
    if c==0 and not pass1[0].isupper():              
         pass1.pop()
         continue
    c=c+1
pass1.append(random.choice(str))
for i in pass1:
     print(i,end="")
```
