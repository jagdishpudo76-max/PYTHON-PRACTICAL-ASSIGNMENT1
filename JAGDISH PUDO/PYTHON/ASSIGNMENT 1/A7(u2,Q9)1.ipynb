```python
import string
text = input("Enter text: ")
print(text)
```

```python
# Convert to lowercase
text = text.lower()
# Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# Split into words
words = text.split()
# Count frequency using dictionary
freq = {}
for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1

print(freq)
```
