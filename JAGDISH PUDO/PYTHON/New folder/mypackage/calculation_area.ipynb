```python

import area
print("Area Options:")
print("Circle\nTriangle\nRectangle\nSquare")
option=input("Eneter area option:")
o=option.lower()
match o:
    case "circle":
        r=int(input("Enter radius:"))
        print(f"Area of circle is {area.circle_area(r)}")
    case "triangle":
        b=int(input("Enter Base:"))
        h=int(input("Enter Height:"))
        print(f"Area of Triangle is {area.triangle_area(b,h)}")
    case "rectangle":
        w=int(input("Enter width:"))
        h=int(input("Enter height:"))
        print(f"Area of Rectangle is {area.rectangle_area(w,h)}")
    case "square":
        h=int(input("Enter Height:"))
        print(f"Area of Square is {area.square_area(h)}")
    
```
